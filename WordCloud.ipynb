```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# You can enter any text
text = "According to a recent whitepaper by Arcserve, there will be over 100 zettabytes of data stored in the cloud by 2025 - the equivalent of around 50% of the worlds data by that time, and a significant increase from the 25% stored in the cloud back in 2015."

cloud = WordCloud().generate(text)
plt.imshow(cloud)
plt.show()
```
