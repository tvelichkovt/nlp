```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt


text = "About 66 million years ago, an estimated 6-mile-wide (9.6 kilometers) object slammed into Earth, triggering a cataclysmic series of events that resulted in the demise of non-avian dinosaurs. Now, scientists think they know where that object came from. According to new research, the impact was caused by a giant dark primitive asteroid from the outer reaches of the solar system's main asteroid belt, situated between Mars and Jupiter. This region is home to many dark asteroids — space rocks with a chemical makeup that makes them appear darker (reflecting very little light) compared with other types of asteroids."

cloud = WordCloud().generate(text)
plt.imshow(cloud)
plt.show()

cloud = cloud.to_file('cloud.png')
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

text = "About 66 million years ago, an estimated 6-mile-wide (9.6 kilometers) object slammed into Earth, triggering a cataclysmic series of events that resulted in the demise of non-avian dinosaurs. Now, scientists think they know where that object came from. According to new research, the impact was caused by a giant dark primitive asteroid from the outer reaches of the solar system's main asteroid belt, situated between Mars and Jupiter. This region is home to many dark asteroids — space rocks with a chemical makeup that makes them appear darker (reflecting very little light) compared with other types of asteroids."

shape = np.array(Image.open("cloud.png"))

cloud = WordCloud(background_color='white', mask = shape).generate(text)

plt.imshow(cloud)
plt.axis('off')
plt.show()

cloud = cloud.to_file('cloud2.png')
```
